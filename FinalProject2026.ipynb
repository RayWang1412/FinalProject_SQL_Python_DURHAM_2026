```python
import datetime #for tha dates and time
import pandas as pd #for DATA FREAMES aka the n.n data
import matplotlib.pyplot as plt #for the plots
import numpy as np #for linear algebra
```

```python
df_Sains = pd.read_csv("All_Data_Sains.csv", header=None)
df_Sains.columns = ['supermarket', 'prices_(£)', 'prices_unit_(£)', 'unit', 'names', 'date', 'category', 'own_brand']
df_Sains = df_Sains.iloc[1:].copy() # Drop the first row which is the actual header
df_Sains['prices_(£)'] = pd.to_numeric(df_Sains['prices_(£)'], errors='coerce')
df_Sains['date'] = pd.to_datetime(df_Sains['date'], format='%Y%m%d', errors='coerce')
df_Sains.dropna(subset=['prices_(£)', 'date'], inplace=True) # Drop rows with NaNs in critical columns
print("Sainsbury's DataFrame (first 5 rows) after cleaning:")
display(df_Sains.head())
df_Sains.info()
```

```python
df_Tesco = pd.read_csv("All_Data_Tesco.csv")
df_Tesco['prices_(£)'] = pd.to_numeric(df_Tesco['prices_(£)'], errors='coerce')
df_Tesco['date'] = pd.to_datetime(df_Tesco['date'].astype(str).str.replace('\.0', '', regex=False), format='%Y%m%d', errors='coerce')
df_Tesco.dropna(subset=['prices_(£)', 'date'], inplace=True) # Drop rows with NaNs in critical columns
print("Tesco DataFrame (first 5 rows) after cleaning:")
display(df_Tesco.head())
df_Tesco.info()
```

## Visualizing Price Trends for 'Food Cupboard' Category

Let's analyze the price trends over time for the 'food_cupboard' category in both Sainsbury's and Tesco.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for 'food_cupboard' category
sains_fc = sains_filtered[sains_filtered['category'] == 'food_cupboard'].copy()
tesco_fc = tesco_filtered[tesco_filtered['category'] == 'food_cupboard'].copy()

# Convert 'date' column to datetime objects if not already converted
sains_fc['date_dt'] = pd.to_datetime(sains_fc['date'], format='%Y%m%d', errors='coerce')
tesco_fc['date_dt'] = pd.to_datetime(tesco_fc['date'], format='%Y%m%d', errors='coerce')

# Calculate daily average prices for food_cupboard category
sains_trend = sains_fc.groupby('date_dt')['prices_(£)'].mean()
tesco_trend = tesco_fc.groupby('date_dt')['prices_(£)'].mean()

# Plotting the trends
plt.figure(figsize=(12, 6))
plt.plot(sains_trend.index, sains_trend.values, label='Sainsbury\'s Food Cupboard', marker='o', markersize=4)
plt.plot(tesco_trend.index, tesco_trend.values, label='Tesco Food Cupboard', marker='x', markersize=4)

plt.title('Average Daily Prices for Food Cupboard Category (Sainsbury\'s vs Tesco)')
plt.xlabel('Date')
plt.ylabel('Average Price (£)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Concatenate Sainsbury's and Tesco dataframes again with corrected df_Sains
# Ensure 'date' column is consistent before concatenation
df_Sains['date'] = pd.to_datetime(df_Sains['date'], format='%Y%m%d', errors='coerce')
df_Tesco['date'] = pd.to_datetime(df_Tesco['date'], format='%Y%m%d', errors='coerce')

combined_df = pd.concat([df_Sains, df_Tesco])

# Drop any rows where price is NaN (e.g., from header row or bad data)
combined_df.dropna(subset=['prices_(£)'], inplace=True)

# Convert 'prices_(£)' to numeric, coercing errors to NaN and then dropping them
combined_df['prices_(£)'] = pd.to_numeric(combined_df['prices_(£)'], errors='coerce')
combined_df.dropna(subset=['prices_(£)'], inplace=True)

# Calculate the average price for each category
average_prices_by_category = combined_df.groupby('category')['prices_(£)'].mean().sort_values(ascending=False)

# Get the top 5 categories with the highest average prices
top_5_categories = average_prices_by_category.head(5)

print("Recalculated Top 5 Categories by Average Selling Price:")
print(top_5_categories)
```

## Time Series Plot for Top 5 Categories by Average Selling Price

Here are the average daily price trends for the top 5 categories based on the corrected data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter combined_df for the top 5 categories
top_5_category_names = top_5_categories.index.tolist()
filtered_top_categories_df = combined_df[combined_df['category'].isin(top_5_category_names)].copy()

# Ensure 'date' column is in datetime format (already done above, but good to be explicit)
# filtered_top_categories_df['date_dt'] = pd.to_datetime(filtered_top_categories_df['date'], errors='coerce')
filtered_top_categories_df.rename(columns={'date': 'date_dt'}, inplace=True)

# Drop rows where date conversion failed (if any)
filtered_top_categories_df.dropna(subset=['date_dt'], inplace=True)

# Calculate daily average prices for each of the top 5 categories
daily_avg_prices_top_categories = filtered_top_categories_df.groupby(['date_dt', 'category'])['prices_(£)'].mean().reset_index()

# Plotting the time series
plt.figure(figsize=(15, 8))
sns.lineplot(data=daily_avg_prices_top_categories, x='date_dt', y='prices_(£)', hue='category', marker='o')

plt.title('Average Daily Prices for Top 5 Categories Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price (£)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
